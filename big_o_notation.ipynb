```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import math

plt.figure(figsize=(14, 10))

n = pd.Series(range(0, 4, 1))

y1 = 0*n + 1
plt.plot(y1, label='O(1)')

y2 = np.log(n)
plt.plot(y2, label='O(log n)')

y3 = n
plt.plot(y3, label='O(n)')

y4 = n * np.log(n)
plt.plot(y4, label='O(n log n)')

y5 = n ** 2
plt.plot(y5, label='O(n^2)')

y6 = 2 ** n
plt.plot(y6, label='O(2^n)')

y7 = n.apply(math.factorial) # coulnt get the math.factorial or my user create factorial to vectorize so had to use the apply() method for this one
plt.plot(y7, label='O(n!)')

y8 = n ** n
plt.plot(y8, label='O(n^n)')

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()

plt.show()
```

```python
# write a function to find the factorial of a number
# factorial of 5 = 5 * 4 * 3 * 2 * 1

def factorial(n):
    for i in range(1 , n):
        n *= i
    return n
```
