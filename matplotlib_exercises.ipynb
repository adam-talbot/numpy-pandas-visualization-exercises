#### 1. Use matplotlib to plot the following equation: y = x**2 - x + 2. You'll need to write the code that generates the x and y points. Add an anotation for the point 0, 0, the origin.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data for x values
x = pd.Series(range(0, 57, 1))

# write function to get y values
def squared_minus_x_plus_two(x):
    return x ** 2 - x + 2
y = x.apply(squared_minus_x_plus_two)
y.head()
# this seems to be a long way to do things when we could just leverage vectorized operation of pandas since our x values are a pandas series
```

```python
# do the same using a lambda to generate data for y values
y = (lambda x: x**2 - x + 2)(x)
y.head()
# shorter, but still not necessary
```

```python
# can't we just do this simply by leveraging vectorized operations in pandas?
y = x**2 - x + 2
y.head()
# same result, shorter and more straightforward
```

```python
plt.plot(y, c = 'green')
plt.annotate('Origin', xy=(0, 0), xytext=(-0, 500),
             arrowprops={'facecolor': 'red'})
plt.show()
```

#### 3. Combine the figures you created in the last step into one large figure with 4 subplots. 

```python
n_rows = 2
n_cols = 2

plt.figure(figsize=(14, 10))

# x values
x = pd.Series(range(0, 57, 1))

# plot the first subplot
plt.subplot(n_rows, n_cols, 1)
y1 = x ** 0.5
plt.plot(y1)
plt.title('y = x**0.5')
plt.xlabel('x')
plt.ylabel('y')

# the second subplot
plt.subplot(n_rows, n_cols, 2)
y2 = x ** 3
plt.plot(y2)
plt.title('y = x**3')
plt.xlabel('x')
plt.ylabel('y')

# the third subplot
plt.subplot(n_rows, n_cols, 3)
y3 = 2 ** x
plt.plot(y3)
plt.title('y=2**x')
plt.xlabel('x')
plt.ylabel('y')

# the fourth subplot
plt.subplot(n_rows, n_cols, 4)
y4 = 1 / (x + 1)
plt.plot(y4)
plt.title('y = 1 / (x+1)')
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()

plt.show()
```

#### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = pd.Series(range(0, 10, 1))

y1 = x ** 0.5
plt.plot(y1, label='y=x**0.5')

y2 = x ** 3
plt.plot(y2, label='y=x**3')

y3 = 2 ** x
plt.plot(y3, label='y=2**x')

y4 = 1 / (x + 1)
plt.plot(y4, label='y=1/(x+1)')

plt.title('Combined Plots')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```
